```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load dataset properly
df = pd.read_csv(r"C:\Users\SSHAR\Downloads\Chocolate Sales (2).csv")

# Step 2: Clean Amount column
df["Amount"] = df["Amount"].replace('[\$,]', '', regex=True).astype(float)

# Step 3: Group by Product
product_sales = df.groupby("Product")["Amount"].sum().sort_values(ascending=False)

# Step 4: Draw Bar Graph
plt.figure()
product_sales.plot(kind="bar")
plt.title("Product Wise Total Sales")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\SSHAR\Downloads\Chocolate Sales (2).csv")

# Clean Amount
df["Amount"] = df["Amount"].replace('[\$,]', '', regex=True).astype(float)

# Convert Date
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)

# Extract Month-Year
df["Month"] = df["Date"].dt.to_period("M")

monthly_sales = df.groupby("Month")["Amount"].sum()

plt.figure()
monthly_sales.plot()
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv(r"C:\Users\SSHAR\Downloads\Chocolate Sales (2).csv")

# Clean Amount column
df["Amount"] = df["Amount"].replace('[\$,]', '', regex=True).astype(float)

# Select numerical columns
numeric_data = df[["Amount", "Boxes Shipped"]]

# Calculate correlation
correlation = numeric_data.corr()

# Plot Heatmap (without seaborn)
plt.figure()
plt.imshow(correlation)
plt.colorbar()
plt.xticks([0,1], correlation.columns)
plt.yticks([0,1], correlation.columns)
plt.title("Correlation Heatmap")
plt.show()

# Print correlation value
print(correlation)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv(r"C:\Users\SSHAR\Downloads\Chocolate Sales (2).csv")

# Clean Amount column
df["Amount"] = df["Amount"].replace('[\$,]', '', regex=True).astype(float)

# Group by Country and Product
data = df.groupby(["Country", "Product"])["Amount"].sum().unstack()

# Plot Stacked Bar Chart
plt.figure()
data.plot(kind="bar", stacked=True)
plt.title("Country Wise Product Sales")
plt.xlabel("Country")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv(r"C:\Users\SSHAR\Downloads\Chocolate Sales (2).csv")

# Clean Amount column
df["Amount"] = df["Amount"].replace('[\$,]', '', regex=True).astype(float)

# Create boxplot country wise
plt.figure()
df.boxplot(column="Amount", by="Country")
plt.title("Country Wise Sales Distribution")
plt.suptitle("")  # remove default subtitle
plt.xlabel("Country")
plt.ylabel("Sales Amount")
plt.xticks(rotation=45)
plt.show()
```
